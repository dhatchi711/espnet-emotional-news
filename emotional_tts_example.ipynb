```python
from IPython.display import display, Audio

from espnet2.sds.tts.espnet_tts import ESPnetTTSModel
from espnet2.sds.tts.emotional_tts import EmotionalTTS

# NOTE: in order to use emotional TTS, you need to install gdown & yacs in addition to the packages required by ESPnet.
# NOTE: pre-trained parameters are downloaded from Google Drive; I did not use ESPnet-zoo because model pull-request may take a time and it is easier to implement.
```

```python
emotion_labels = ["Neutral", "Happy", "Sad", "Angry", "Surprise"]
content = "I am a student at Carnegie Mellon University."
```

```python
tts = ESPnetTTSModel(device="cpu")
emotional_tts = EmotionalTTS(device="cpu") # OPTION: you can choose speaker from 0011, 0012, ..., 0020 by adding an argument "speaker" (see espnet2.sds.tts.emotional_tts.py for more details)
```

```python
transcript = f"<{emotion_labels[0]}>{content}"
emotional_tts.split_emotion_label_transcript(transcript)
```

```python
fs, audio_chunk = tts.forward(content)
display(Audio(audio_chunk.T, rate=fs))
```

```python
for emotional_label in emotion_labels:
    transcript = f"<{emotional_label}>{content}"

    print(transcript)
    fs, audio_chunk = emotional_tts.forward(transcript)
    display(Audio(audio_chunk.T, rate=fs))
    print("------")
```
